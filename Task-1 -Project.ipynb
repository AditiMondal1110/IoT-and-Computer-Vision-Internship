**Importing necessary libraries**


```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import display,Image
```

```python
# Image  before OCR
image_path = "car3.jpg"
im=cv2.imread('car3.jpg')
plt.imshow(im)
plt.show()
```

```python
reader=easyocr.Reader(['en'],gpu=False)
result=reader.readtext(image_path)
result
```

```python
top_left=result[0][0][0]
bottom_right=result[0][0][2]
text=result[0][1]
Font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
new_top_left=[]
for i in top_left:
    new_top_left.append(i+25)
```

```python
img=cv2.imread(image_path)
img=cv2.rectangle(img,top_left,bottom_right,(0,255,0),10)
img=cv2.putText(img,text,new_top_left,Font,2,(255,0,0),2)
plt.figure(figsize= (7,7))
plt.imshow(img)
plt.show()
```
